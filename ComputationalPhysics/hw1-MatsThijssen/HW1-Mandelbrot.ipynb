```python
import numpy as np
#from pylab import imshow,show, gray
import matplotlib.pyplot as plt
```

```python
Data=np.empty([2001,2001]) #Initialize data-array




z_real=0
z_imag=0

step=4/2000 #Step size
#Loop through different values for c (x real part, y imaginary part) and check if they hold up through 100 iterations.

for x in np.arange(-2.4,2+step,step):
    for y in np.arange(-2.4,2+step,step):
        fail='no'
        z_real=0
        z_imag=0
        zmag=z_real**2+z_imag**2
        zsqreal=z_real**2-z_imag**2
        zsqimag=2*z_real*z_imag
        for i in range(0,300):
            z_real=zsqreal+x
            z_imag=zsqimag+y
            zmag=z_real**2+z_imag**2
            zsqreal=z_real**2-z_imag**2
            zsqimag=2*z_real*z_imag
            if abs(zmag)>2:
                fail='yes' #If they don't hold, fail them.
                failnr=i
                break
        if fail=='yes':
            Data[(2+y)/step,(2+x)/step]=(failnr/300)-0.05 #If they fail, give the point value '0'
        else:
            Data[(2+y)/step,(2+x)/step]=1 #If they succeed, give the point value 1
```

```python
#Plot!
plt.figure(figsize=(20,20), dpi=100)
plt.imshow(Data, origin='lower')
plt.set_cmap('magma')
plt.show()
```
