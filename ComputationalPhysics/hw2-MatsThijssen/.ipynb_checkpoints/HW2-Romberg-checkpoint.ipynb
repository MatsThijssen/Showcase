```python
import numpy as np
from numba import *
import gaussxw as gs
import matplotlib.pyplot as plt
```

```python
def f1(x):
    return (x**2) * np.cos(x)
def f2(x):
    return x**(1/2)


##min
#b=4*np.pi #max
#h=b-a #step size
#N=20 #max steps
@jit
def Romberg(a,b,f,N):
    h=b-a
    R=np.empty([N+1,N+1])
    R[0,0]=(h/2)*(f(a)+f(b))

    for n in range(1,N):
        Sum_thing=0
        for k in np.arange(1, 2**(n-1)+1):
            Sum_thing+=f(a+(2*k-1)*(h/2**n))
        R[n,0]=(1/2) *R[n-1,0]+(h/2**n)*Sum_thing
    for i in range(1,N):
        for m in range(1,N):
            R[i,m]=(1/((4**m) -1))*((4**m)*R[i,m-1] -R[i-1,m-1] )
    return R[N-1,N-1]
```

```python
@jit
def Gauss(a,b,f,N):
    x,w=gs.gaussxwab(N,a,b)
    s=0
    for k in range(0,N):
        s+=w[k]*f(x[k])
    return s
        
    
```

```python
act1=8*np.pi
act2=200*np.pi
act3=2/3
Romerr1=np.empty([20])
Romerr2=np.empty([20])
Romerr3=np.empty([20])
for n in range(1,21):
    Romerr1[n-1]=abs(act1-Romberg(0,4*np.pi,f1,n))
    Romerr2[n-1]=abs(act2-Romberg(0,100*np.pi,f1,n))
    Romerr3[n-1]=abs(act3-Romberg(0,1,f2,n))
```

```python
act1=8*np.pi
act2=200*np.pi
act3=2/3
N=200
Gausserr1=np.empty([N])
Gausserr2=np.empty([N])
Gausserr3=np.empty([N])
for n in range(1,N+1):
    Gausserr1[n-1]=abs(act1-Gauss(0,4*np.pi,f1,n))
    Gausserr2[n-1]=abs(act2-Gauss(0,100*np.pi,f1,n))
    Gausserr3[n-1]=abs(act3-Gauss(0,1,f2,n))
```

```python




plt.figure(figsize=(21,5),dpi=150)
for i in range(1,200):
    plt.subplot(1,3,1)
    plt.title('Gauss Error plot for function 1' )
    plt.xlabel('ln(n)')
    plt.ylabel('ln(L1)')
    plt.plot(np.log(i),np.log(Gausserr1[i-1]),'r+')
    plt.subplot(1,3,2)
    plt.title('Gauss Error plot for function 2' )
    plt.xlabel('ln(n)')
    plt.ylabel('ln(L1)')
    plt.plot(np.log(i),np.log(Gausserr2[i-1]),'r+')
    plt.subplot(1,3,3)
    plt.title('Gauss Error plot for function 3' )
    plt.xlabel('ln(n)')
    plt.ylabel('ln(L1)')
    plt.plot(np.log(i),np.log(Gausserr3[i-1]),'r+')
plt.show()
```

```python
#Slope of fit
pts=np.linspace(1,100,100)
m,b = np.polyfit(np.log(2**pts[:100]), np.log(Gausserr3[:100]), 1)
print(m,b)
```
