```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
```

```python
#Initial stuff
v=1 #speed
delta_t=0.01
N=10000
delta_x=40/N


u=np.empty([N+1])
u_storage=np.empty([N+1])
prop_gauss=np.empty([N+1])
x=np.empty([N+1])
y=0
for i in np.arange(-20,20+0.0000001,delta_x):
    x[y]=i
    u[y]=0.5*np.exp(-(i**2)/0.1)
    prop_gauss[y]=1*np.exp(-((i-v*64)**2)/500)
    y+=1
u_storage=u
plt.figure(figsize=(5,5),dpi=200)
plt.plot(x,u,'b-')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python


for t in np.arange(0,8,delta_t):
    for i in range(1,500):
        u[i]=u_storage[i]-((v*delta_t)/(2*delta_x))*(u_storage[i+1]-u_storage[i-1])
    u_storage=u
    #if (t/0.2)%1==0:
plt.figure(figsize=(5,5),dpi=200)
plt.plot(x,prop_gauss, 'r--')
plt.plot(x,u, 'b-')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
for t in np.arange(0,16,delta_t):
    for i in range(1,N):
        u[i]=(0.5*(u_storage[i+1]+u_storage[i-1]))-(((v*delta_t)/(2*delta_x))*(u_storage[i+1]-u_storage[i-1]))
    u_storage=u
    #if (t/0.2)%1==0:
plt.figure(figsize=(5,5),dpi=200)
plt.plot(x,prop_gauss, 'r--')
plt.plot(x,u, 'b-')
plt.title("t=16")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
