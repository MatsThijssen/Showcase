```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 1D SR Hydro code

class Relativistic_Hydro:
    
    Nx = 10
    a = 0.0
    b = 1.0
    t = 0.0
    cfl = 0.01
    Gamma = 1.4
    dx = 0.1
    x = None
    p = None
    rho = None
    v = None
    
    def __init__(self, Nx, a, b, t, cfl, Gamma):
        
        # scalars
        self.Nx = Nx
        self.a = a
        self.b = b
        self.t = t
        self.cfl = cfl
        self.Gamma = Gamma
        self.dx = (b-a) / float(Nx)
        
        # arrays
        self.x = a + self.dx*(np.arange(Nx)+0.5)
        self.p = np.zeros(Nx)
        self.rho = np.zeros(Nx)
        self.v = np.zeros(Nx)
    
    def setInitialConditions(self, f):
        self.p = f(self.x, 'p')
        self.rho = f(self.x, 'rho')
        self.v = f(self.x, 'v')
        
    def evolve(self, tfinal):
        
        while self.t < tfinal:
            
            # primitive variables
            p = np.array(self.p)               # pressure 
            rho = np.array(self.rho)           # density
            v = np.array(self.v)               # velocity
            
            # intermediary quantities
            Gamma = self.Gamma                 # adiabatic index
            W = np.sqrt(1/(1-v**2))            # Lorentz factor
            eps = (1/(Gamma-1)) * p / rho      # specific internal energy
            h = 1 + eps + p/rho                # specific enthalpy
            
            # quantities needed for flux
            cs = np.sqrt(Gamma*p / (rho*h))    # sound speed
            lp = (v + cs)/(1 + v*cs)           # lambdas, using relativistic velocity addition
            lm = (v - cs)/(1 - v*cs)
            ap = np.array([max(0, lp[i], lp[i+1]) for i in range(self.Nx-1)])
            am = np.array([max(0,-lm[i],-lm[i+1]) for i in range(self.Nx-1)])
            
            # conserved variables
            D = rho*W                          # lab frame density
            S = rho*h*W*W*v                    # lab frame momentum density
            tau = rho*h*W*W - p - rho*W        # lab frame non-rest-mass energy
            
            U = np.zeros((3,self.Nx))
            U[0] = D
            U[1] = S
            U[2] = tau
            
            # fluxes
            F = np.zeros((3,self.Nx))
            F[0] = D*v
            F[1] = S*v + p
            F[2] = S - D*v
            
            # calculate and update time step
            v_for_dt = np.fabs(np.concatenate(([0],ap,am))).max()
            #if v_for_dt > 3e8:
            #    v_for_dt = 3e8
            dt = self.cfl * self.dx / v_for_dt
            #dt = 1e-4
            if self.t + dt > tfinal:
                dt = tfinal - self.t
            self.t += dt
            
            # HLL flux
            FL = F[:,:-1]
            FR = F[:,1:]
            UL = U[:,:-1]
            UR = U[:,1:]
            FHLL = (ap*FL + am*FR - ap*am*(UR-UL)) / (ap+am)
            
            # calculate L(U)
            LU = np.zeros((3,self.Nx))
            LU[:,1:-1] = -(FHLL[:,1:] - FHLL[:,:-1]) / self.dx
            
            # update conserved variables
            U[:,1:-1] += dt*LU[:,1:-1]
            
            # conversion back to primitive variables
            D = U[0]
            S = U[1]
            tau = U[2]
            p = self.p
            v = self.v
            a = 0
            b = 0
            
            
            for i in range(5):       # Newton-Raphson for finding p
                
                rhoNR = D*np.sqrt((tau+p+D)**2-S**2)/(tau+p+D)
                eNR = ((np.sqrt((tau+p+D)**2-S**2)/D)-((p/rhoNR)+1.0))
                dpdrho = (Gamma-1)*eNR
                dpde = (Gamma-1)*rhoNR
                drhodp = D*S**2/(np.sqrt((tau+p+D)**2-S**2)*(tau+p+D)**2)
                dedp = p*S**2/(rhoNR*((tau+p+D)**2-S**2)*(tau+p+D))
            
                f = p - (Gamma-1)*rhoNR*eNR
                der = 1.0 - dpdrho*drhodp - dpde*dedp
                
                for j in range(self.Nx):
                    if ((tau[j]+D[j])**2) > D[j]**2+S[j]**2:
                        if (der[j]!=0.0):
                            p[j] -= f[j]/der[j]
                
            self.p = p
            self.rho = rhoNR
            h = 1 + eNR + p/rhoNR
                
            for j in range(self.Nx):
                if (np.fabs(S[j])!=0):
                    self.v[j] = (-(rhoNR[j]*h[j])+np.sqrt((rhoNR[j]*h[j])**2+(2*S[j])**2))/(2*S[j])

    def plot(self, ax=None, filename=None):
        if ax is None:
            fig, ax = plt.subplots(3, sharex=True, figsize=(5,10))
        else:
            fig = ax.get_figure()
        
        ax[0].set_title("t = "+str(self.t), fontsize=24)
        ax[0].plot(self.x, self.rho, 'k')
        ax[0].set_ylabel('rho', fontsize=16)
        ax[1].plot(self.x, self.v, 'k')
        ax[1].set_ylabel('v', fontsize=16)
        ax[2].plot(self.x, self.p, 'k')
        ax[2].set_ylabel('p', fontsize=16)
        ax[2].set_xlabel('x', fontsize=16)
        plt.tight_layout()
        
        if filename is not None:
            fig.savefig(filename)
            
        return ax
```

```python
# Sod problem

H = Relativistic_Hydro(400, 0.0, 1.0, 0.0, 0.4, 1.4)
def Sod(x, var):
    Sod = np.zeros(len(x))
    for i in range(len(x)):
        if x[i]<0.5:
            p0 = 13.33
            rho0 = 10.0
            v0 = 0.001
        else:
            p0 = 1e-8
            rho0 = 1.0
            v0 = 0.001
        if var == 'p':
            Sod[i] = p0
        if var == 'rho':
            Sod[i] = rho0
        if var == 'v':
            Sod[i] = v0
    return Sod
H.setInitialConditions(Sod)
#H.plot(filename="Sod_t0.pdf")
H.evolve(0.4)
H.plot()
```
